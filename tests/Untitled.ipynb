```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import os
import re
from pyspark.sql.types import *
from pyspark.sql.functions import udf,coalesce,col,regexp_replace,split,trim,struct
input_path = "/home/luizlobo/Documents/code/imoveis/tests/samples/data"
df = spark.read.json(input_path)
```

```python
regexp_non_words = re.compile(r"^\W+|\W+$",flags=re.UNICODE)
# Remove spaces without afecting accent words
def strip_with_utf8(string):
    print(string)
```

```python
# First treat the data in strings removing starting spaces unused
str_columns = [column_name for column_name,data_type in df.dtypes if data_type == 'string']
for column in str_columns:
    df = df.withColumn(column,regexp_replace(strip_with_utf8(column),r"\s+"," "))
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
replace_utf8 = udf(lambda newString: strip_with_utf8(newString), StringType())
replace_utf8(df["bairro"])
df.withColumn('bairro2',replace_utf8(col("bairro"))).select("bairro2").show(20,False)
```
